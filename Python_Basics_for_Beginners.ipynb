## 1. Variables and Data Types

A **variable** is a container for storing data values. Python is dynamically typed, meaning you don't need to declare the type—Python figures it out!

```python
# Variables and assignment
name = "Alice"           # String (text)
age = 25                 # Integer (whole number)
height = 5.6             # Float (decimal number)
is_student = True        # Boolean (True or False)

print(name)
print(age)
print(height)
print(is_student)
```

```python
# Check the type of a variable
print(type(name))
print(type(age))
print(type(height))
print(type(is_student))
```

```python
# String operations
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name  # Concatenation
print(full_name)

# String methods
print(first_name.upper())     # Convert to uppercase
print(first_name.lower())     # Convert to lowercase
print(len(first_name))        # Length of string
```

## 2. Operators

Operators are used to perform operations on variables and values.

```python
# Arithmetic operators
a = 10
b = 3

print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")
print(f"Division: {a} / {b} = {a / b}")
print(f"Floor Division: {a} // {b} = {a // b}")
print(f"Modulo (remainder): {a} % {b} = {a % b}")
print(f"Exponent: {a} ** {b} = {a ** b}")
```

```python
# Comparison operators
x = 5
y = 8

print(f"{x} == {y}: {x == y}")   # Equal to
print(f"{x} != {y}: {x != y}")   # Not equal to
print(f"{x} > {y}: {x > y}")     # Greater than
print(f"{x} < {y}: {x < y}")     # Less than
print(f"{x} >= {y}: {x >= y}")   # Greater than or equal to
print(f"{x} <= {y}: {x <= y}")   # Less than or equal to
```

```python
# Logical operators
a = True
b = False

print(f"{a} and {b}: {a and b}")   # Both must be True
print(f"{a} or {b}: {a or b}")     # At least one must be True
print(f"not {a}: {not a}")          # Reverses the boolean
```

## 3. Control Flow: if/else/elif

Control flow allows you to execute different code based on conditions.

```python
# Simple if/else
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
else:
    print("Grade: F")
```

```python
# Nested conditions
age = 25
has_license = True

if age >= 18:
    if has_license:
        print("You can drive!")
    else:
        print("You need a driver's license")
else:
    print("You are too young to drive")
```

## 4. Loops

Loops allow you to repeat code multiple times.

```python
# for loop with range()
print("Counting from 1 to 5:")
for i in range(1, 6):  # range(start, end) - end is exclusive
    print(i)
```

```python
# for loop with a list
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(f"I like {fruit}")
```

```python
# while loop
count = 0

while count < 5:
    print(f"Count: {count}")
    count += 1  # count = count + 1
```

```python
# break and continue
print("Using break:")
for i in range(10):
    if i == 5:
        break  # Exit the loop
    print(i)

print("\nUsing continue:")
for i in range(5):
    if i == 2:
        continue  # Skip this iteration
    print(i)
```

## 5. Functions

Functions are reusable blocks of code that perform a specific task.

```python
# Simple function
def greet():
    print("Hello, World!")

greet()  # Call the function
```

```python
# Function with parameters
def greet_person(name):
    print(f"Hello, {name}!")

greet_person("Alice")
greet_person("Bob")
```

```python
# Function with return value
def add(a, b):
    result = a + b
    return result

sum_result = add(5, 3)
print(f"5 + 3 = {sum_result}")
```

```python
# Function with default parameters
def introduce(name, age=25):
    print(f"{name} is {age} years old")

introduce("Alice")           # Uses default age
introduce("Bob", 30)         # Provides custom age
```

```python
# Function with multiple return values
def get_min_max(numbers):
    minimum = min(numbers)
    maximum = max(numbers)
    return minimum, maximum

my_list = [3, 7, 2, 9, 1]
min_val, max_val = get_min_max(my_list)
print(f"Min: {min_val}, Max: {max_val}")
```

## 6. Lists and Dictionaries

Data structures for storing collections of data.

```python
# Lists - ordered, mutable collections
numbers = [1, 2, 3, 4, 5]
print(f"List: {numbers}")

# Accessing elements (indexing starts at 0)
print(f"First element: {numbers[0]}")
print(f"Last element: {numbers[-1]}")

# Slicing
print(f"Elements 1-3: {numbers[1:4]}")
```

```python
# List methods
fruits = ["apple", "banana", "cherry"]

fruits.append("date")        # Add to end
print(f"After append: {fruits}")

fruits.insert(1, "blueberry") # Insert at position
print(f"After insert: {fruits}")

fruits.remove("banana")      # Remove specific item
print(f"After remove: {fruits}")

print(f"Length: {len(fruits)}")
```

```python
# Dictionaries - key-value pairs
person = {
    "name": "Alice",
    "age": 25,
    "city": "New York"
}

print(f"Dictionary: {person}")

# Accessing values
print(f"Name: {person['name']}")
print(f"Age: {person['age']}")
```

```python
# Dictionary operations
person = {"name": "Bob", "age": 30}

# Add a new key-value pair
person["email"] = "bob@example.com"
print(f"After adding email: {person}")

# Update a value
person["age"] = 31
print(f"After updating age: {person}")

# Remove a key
del person["email"]
print(f"After deleting email: {person}")

# Check if key exists
print(f"Has 'name': {'name' in person}")
print(f"Has 'email': {'email' in person}")
```

```python
# Iterating through dictionaries
student = {"name": "Charlie", "grade": "A", "score": 95}

# Get keys
print("Keys:", student.keys())

# Get values
print("Values:", student.values())

# Get key-value pairs
print("Items:")
for key, value in student.items():
    print(f"  {key}: {value}")
```

### Exercise 1: Temperature Converter
A function takes an input, does a calculation, and returns an output. Here, we will use the formula for converting Celsius to Fahrenheit: F = (C × 9/5) + 32.

```python
def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9 / 5) + 32
    return fahrenheit

print(celsius_to_fahrenheit(0))    # 32
print(celsius_to_fahrenheit(100))  # 212
print(celsius_to_fahrenheit(25))   # 77
```

### Exercise 2: Find the Sum
A loop helps us repeat the same action many times. In this exercise, we will add every number in a list together.

```python
def sum_numbers(numbers):
    total = 0
    for number in numbers:
        total = total + number
    return total

print(sum_numbers([1, 2, 3, 4, 5]))  # 15
print(sum_numbers([10, 20, 30]))     # 60
```

### Exercise 3: Dictionary Counter
A dictionary stores data as key-value pairs. Here, each letter becomes a key and the count becomes the value.

```python
def count_letters(text):
    counts = {}

    for char in text.lower():
        if char.isalpha():
            counts[char] = counts.get(char, 0) + 1

    return counts

result = count_letters("hello")
print(result)  # {'h': 1, 'e': 1, 'l': 2, 'o': 1}

result_2 = count_letters("Hello, World!")
print(result_2)
```

### Exercise 4: Filter Even Numbers
We can use an if statement inside a loop to choose only the values that match a condition.

```python
def get_even_numbers(numbers):
    even_numbers = []

    for number in numbers:
        if number % 2 == 0:
            even_numbers.append(number)

    return even_numbers

print(get_even_numbers([1, 2, 3, 4, 5, 6, 7, 8]))  # [2, 4, 6, 8]
```

### Exercise 5: Rock-Paper-Scissors Game
This game combines user input, random choices, and conditional logic. The computer chooses randomly, and we compare both choices to decide the winner.

```python
import random

choices = ["rock", "paper", "scissors"]


def play_game():
    user_choice = input("Choose rock, paper, or scissors: ").strip().lower()

    if user_choice not in choices:
        print("Invalid choice. Please choose rock, paper, or scissors.")
        return

    computer_choice = random.choice(choices)
    print(f"You chose: {user_choice}")
    print(f"Computer chose: {computer_choice}")

    if user_choice == computer_choice:
        print("It's a tie!")
    elif (user_choice == "rock" and computer_choice == "scissors") or \
         (user_choice == "paper" and computer_choice == "rock") or \
         (user_choice == "scissors" and computer_choice == "paper"):
        print("You win!")
    else:
        print("Computer wins!")

#Uncomment the next line if you want to play the game now
play_game()
```
